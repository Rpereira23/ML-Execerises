## Logistic Regression

Given the empirical loss (the risk) for logistuc regression: 

$ J(\theta) = \frac{1}{m} \sum_{i=1}^{m} \mathrm{log}\big(1 + e ^{-y^{(i)} \theta^T x^{(i)}}\big) = - \frac{1}{m} \sum_{i=1}^{m} \mathrm{log}(h_\theta (y^{(i)}x^{(i)}))$

where $ y(i) ∈ \{−1,1\}, h_θ(x) = g(θ^Tx) $ and $ g(z) = 1/(1+e^{−z}) $. Find the Hessian H of this function and show that it is positive semidefinite. 

First, note that:

$\frac{\partial g(z)}{\partial z} \frac{1}{(1+e^{−z})} = \frac{(e^{-z})}{(1+e^{-z})^2} = \frac{(e^{-z})}{(1+e^{-z})} \frac{1}{(1+e^{-z})} = \frac{(e^{-z})}{(1+e^{-z})} g(z) = (1 - g(z))g(z) $


Therefore,

$ \frac{\partial J(\theta)}{\partial \theta_{i}} = - \frac{1}{m} \sum_{i=1}^{m} \frac{1}{g(z)} \frac{\partial g(z)}{\partial z}(y^{(i)}x^{(i)}) = - \frac{1}{m} \sum_{i=1}^{m} \frac{1}{g(z)} (1 - g(z))(g(z)) (y^{(i)}x^{(i)}) = - \frac{1}{m} \sum_{i=1}^{m} (1 - g(z))(y^{(i)}x^{(i)}) $

So, 

$ H_{ij} = - \frac{1}{m} \sum_{k=1}^{m} \frac{\partial J(\theta)}{\partial \theta_{i} \theta_{j}} = (1 - g(z))(y^{(i)}x^{(i)}) = - \frac{1}{m} \sum_{k=1}^{m} (1 - g(z))g(z)(y^{(k)}x_i^{(k)})(y^{(k)}x_j^{(k)}) $ 

Since, $ y^{(i)} ∈ \{−1,1\} $, 

$ H_{ij} = -\frac{1}{m} \sum_{k=1}^{m} (1 - g(z))g(z)x_i^{(k)}x_j^{(k)} $

Lastly, 

\begin{align*}
z^{T}Hz = z^{T} ( \frac{1}{m} \sum_{k=1}^{m} (g(z) - 1)g(z)xx^{T} ) z \\
= \frac{1}{m} \sum_{k=1}^{m} (g(z) - 1)g(z)z^{T}xx^{T}z \\
= \frac{1}{m} \sum_{k=1}^{m} (g(z) - 1)g(z)(x^{T}z)^{2}
\end{align*}

since, $ (g(z) - 1)g(z) ∈ [0, 1] $ and $ (x^{T}z)^{2} >= 0 $, $ H $ is positive semidefinite
