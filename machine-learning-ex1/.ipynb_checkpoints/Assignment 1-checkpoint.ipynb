```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ex1.plotData import plot_data
from ex1.computeCost import compute_cost
from ex1.gradientDescent import gradient_descent
```

```python
# ===================== Part 1: Plotting =====================
```

```python
data = pd.read_csv('ex1/ex1data1.txt', header=None)
data.columns = ['population', 'profit' ]
```

```python
plt.ion()
plt.figure(0)
plot_data(data['population'], data['profit'], 'Population of City in 10,000s', 'Profit in $10,000s')
```

```python
y = data['profit'].values
m = y.size
alpha = 0.01
iterations = 1500
```

```python
X = np.c_[np.ones(m), data['population'].values] #now dim(len(x), 2)
theta = np.zeros(2) # dim(2, 1)
```

```python
compute_cost(X, y, theta)
```

```python
theta, J_history = gradient_descent(X, y, theta, alpha, iterations)
```

```python
plt.figure(0)
line1, = plt.plot(X[:, 1], np.dot(X, theta), label='Linear Regression')
plt.legend(handles=[line1])
```
